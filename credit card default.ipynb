```python
df['EDUCATION'] = df['EDUCATION'].replace([5, 6], 4)
col = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[col] = df[col].replace(0, -1)
```

```python
df = df.rename(columns = {'default.payment.next.month': 'def_pay', 'PAY_0': 'PAY_1'})
df.head()
```

```python
df = df.reset_index(drop = True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.metrics import roc_auc_score,roc_curve,scorer, auc
from sklearn.metrics import precision_score,recall_score, f1_score
```

```python
y = df['def_pay'].copy()
X=df.drop(['ID'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
log = LogisticRegression()
log.fit(X_train, y_train)
y_pred = log.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
### use pca to check how the 2 classes are distributed
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)

def plot_2d_space(X, y, label='Classes'):   
    colors = ['#1F77B4', '#FF7F0E']
    markers = ['o', 's']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==l, 0],
            X[y==l, 1],
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
    

plot_2d_space(X_pca, y_train, 'Imbalanced dataset (2 PCA components)')
```
